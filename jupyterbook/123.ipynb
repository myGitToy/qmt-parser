# 说明
本模块演示了指定的证券代码和基准指数在某个日期区间内的比较。

## 输入参数
- 证券代码（如：'600519'）
- 基准指数（如：'000300'）
- 日期区间（如：'2023-01-01' 至 '2023-12-31'）

## 输出结果
- 证券与基准指数的收益率对比图
- 相关性分析数据表

## 使用场景
适用于量化投资、策略回测、绩效评估等金融分析任务。

## 常用分析指标
### β系数(Beta)

衡量个股相对于市场的波动敏感度
β>1表示波动大于市场，β<1表示波动小于市场
### α系数(Alpha)

衡量个股相对于预期回报的超额收益
正α表示跑赢市场，负α表示跑输市场
相关系数(Correlation)

皮尔逊相关系数：测量线性相关性
斯皮尔曼相关系数：适用于非线性关系
滚动相关系数：观察相关性随时间变化
风险调整收益指标

夏普比率(Sharpe Ratio)：超额收益与波动率比值
特雷诺比率(Treynor Ratio)：超额收益与系统性风险比值
信息比率(Information Ratio)：主动收益与主动风险比值
风险指标

最大回撤(Maximum Drawdown)
下行风险(Downside Risk)
跟踪误差(Tracking Error)
波动性分析

历史波动率(Historical Volatility)
条件异方差模型(GARCH)
